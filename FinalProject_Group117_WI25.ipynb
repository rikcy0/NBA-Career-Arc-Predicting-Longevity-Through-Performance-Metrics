# Abstract

This study examines which advanced statistical metrics best predict NBA career longevity over the past 20 seasons. While traditional statistics like points per game and rebounds provide basic performance insights, modern analytics offer a more comprehensive view of long-term player impact. By analyzing Player Efficiency Rating (PER), Box Plus-Minus (BPM), True Shooting Percentage (TS%), and Win Shares, we identify the key factors influencing career length and positional differences in player longevity.

Through exploratory data analysis and statistical modeling on historical player data, we find that Win Shares and TS% exhibit strong correlations with career longevity. Guards tend to have longer careers when they excel in TS%, while forwards and centers benefit more from defensive-based metrics. These findings higlight the importance of efficiency and contribution to a team's success in sustaining a professional basketball career.

These results show how efficiency metrics aid in predicting career longevity and suggest that different playing styles contribute to different levels of sustained success. Expanding the analysis to include external factors like injuries and roster changes could further enhance our understanding of career trajectories.

Let's make sure to address all the outliers in the data now. In the winshares and career length categories we decided that outliers could be attributed to superstar players or those with long careeers respectively, so we will leave them alone. For true shooting percentage (TS%), we decided to treat anything higher than a 0.7324 TS% an outlier since the highest ever recorded TS% was 0.7324. For the same reason, we treat player efficiency ratings higher than 40.0 as outliers. After these considerations, all the other outliers were found and cleaned with standard calculations. Now lets look at the same histograms one more time.

```python
df["TS%"] = pd.to_numeric(df["TS%"], errors="coerce")
df = df.dropna(subset=["TS%"])

df["BPM"] = pd.to_numeric(df["BPM"], errors="coerce")
df = df.dropna(subset=["BPM"])

#bounds of box plus minus
Q1a = df["BPM"].quantile(0.25)  
Q3a = df["BPM"].quantile(0.75)  
IQRa = Q3a - Q1a

lower_bound_a = Q1a - 1.5 * IQRa
upper_bound_a = Q3a + 1.5 * IQRa

#bounds of player efficiency
Q1b = df["PER"].quantile(0.25)  
Q3b = df["PER"].quantile(0.75)  
IQRb = Q3b - Q1b

lower_bound_b = Q1b - 1.5 * IQRb
upper_bound_b = Q3b + 1.5 * IQRb

df_outliers = df[(df["TS%"] > 0.7324) | (df["TS%"] == 0) |
                ((df["PER"] < lower_bound_b) | (df["PER"] > 40.0)) | 
                ((df["BPM"] < lower_bound_a) | (df["BPM"] > upper_bound_a))]

df_cleaned = df[~df.index.isin(df_outliers.index)].copy()
```

```python
variables2 = ['WS', 'TS%', 'PER', 'BPM', 'DWS', 'career_length']
actual_names2 = ['Win Shares', 'True Shooting Percentage', 'Player Efficiency Rating', 
                'Box Plus Minus', 'Defensive Win Shares', 'Career Length']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
sns.set_palette('muted')
sns.set_context('notebook')

for ax, var, name in zip(axes.flatten(), variables2, actual_names2):
    sns.histplot(data=df_cleaned, x=var, kde=True, ax=ax)
    stats = df_cleaned[var].describe().round(2)
    if var != 'TS%':
        ax.text(0.95, 0.95, stats.to_string(), transform=ax.transAxes,
                fontsize=10, verticalalignment='top', horizontalalignment='right',
                bbox=dict(boxstyle="round", facecolor="white", alpha=0.5))
    else:
        ax.text(0.05, 0.95, stats.to_string(), transform=ax.transAxes,
                fontsize=10, verticalalignment='top', horizontalalignment='left',
                bbox=dict(boxstyle="round", facecolor="white", alpha=0.5))
    ax.set_title(f'Distribution of {name}')

plt.tight_layout()
plt.show()
```

As briefly mentioned before, predicting career longevity from a combination of different metrics is suitable for a regression approach — namely multiple linear regression. By treating career length as the outcome variable, multiple regression allows us to include multiple predictors and estimate how much each metric correlates to a player's overall expected career length, which is returned to us as the coefficients of our predictor variables. Additionally, assuming linearity in the parameters allows us to interpret the coefficients easier, as each weight will represent how much a one-unit change in the associated metric affects the predicted career length (assuming all the other variables are held constant).

However, when training a regression model, it is important to be mindful of the presence of multicollinearity in our features, which occurs when two or more predictors are highly correlated with each other. When multicollinearity is present, it will introduce extreme variance in the resulting coefficients, which will reduce interpretability of our model overall. A correlation matrix can show us a quick overview of all the pairwise relationships and their corresponding correlation coefficients for all the numeric features, which will let us identify which variables have the greatest correlations with career length, as well as detecting multicollinearity to see if two variables are highly correlated. For exploratory purposes, it might be beneficial to include a couple of other variables to see if they would be suitable. We'll examine value over replacement player (VORP), minutes played during a season (MP), and turnover percentage (TOV%).

A good starting point for our regression model is to choose predictors that correlate well with our target variable, career length, and do not exhibit high levels of multicollinearity.

```python
variables_of_interest = ['WS', 'TS%', 'PER', 'BPM', 'DWS', 'VORP', 'MP', 'TOV%', 'career_length']

corr_matrix = df_cleaned[variables_of_interest].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Key Variables")
plt.show()
```

From this correlation matrix, we can see that in descending order, the variables that have the highest magnitude of correlation with career length are minutes played, winshares, defensive winshares, value over replacement player, and box plus minus. These are the variables that we might want to include as parameters in our model. However, we can see that WS is highly correlated with DWS, VORP, and MP, with correlation scores of over 0.83, which would lead to concerns over multicollinearity. Similarly, there is a high correlation between player efficiency rating and box plus minus, as well as between defensive win shares and minutes played. It could be beneficial to exclude win shares and keep defensive win shares, since they have similar levels of correlation with career length and defensive win shares is less correlated with the other variables.

Despite the fact that there might be collinearity between these predictor variables, it can still be acceptable to use them as parameters in our model if we apply techniques like dimensionality reduction or regularization. These techniques can help mitigate the impact of highly correlated features in our model. Otherwise, it might be too difficult to find a set of parameters that aren't highly correlated because a lot of them naturally are due to how these statistics are recorded, as well as the nature of basketball itself.

We will also display the weights that each variable had on predicting the career length in order to determine which statistic is the most important.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.svm import SVC
from sklearn.metrics import classification_report, precision_recall_fscore_support
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import classification_report
```

```python
def classify(years):
    if years <= 5.0:
        return 0
    elif years <= 10.0:
        return 1
    elif years <= 15.0:
        return 2
    else:
        return 3

comb_line_X = df_cleaned[['WS', 'TS%', 'PER', 'BPM', 'DWS', 'VORP', 'MP', 'TOV%']]
comb_line_Y = df_cleaned['career_length']

comb_line_train_X, comb_line_test_X, comb_line_train_Y, comb_line_test_Y = train_test_split(
    comb_line_X,
    comb_line_Y,
    test_size = 0.2,
    random_state = 42
)

scaler = StandardScaler()
comb_line_train_X = scaler.fit_transform(comb_line_train_X)
comb_line_test_X = scaler.transform(comb_line_test_X)

def train_SVM(X, y, kernel = 'linear'):
    clf = SVC(kernel = kernel)
    clf.fit(X, y)
    return clf

comb_line_clf = train_SVM(comb_line_train_X, comb_line_train_Y)

comb_line_pred_train_Y = comb_line_clf.predict(comb_line_train_X)
comb_line_pred_test_Y = comb_line_clf.predict(comb_line_test_X)

comb_line_train_Y_clf = comb_line_train_Y.apply(classify)
comb_line_pred_train_Y = [classify(y) for y in comb_line_pred_train_Y]
class_report_train = classification_report(comb_line_train_Y_clf, comb_line_pred_train_Y, target_names=["0.0-5.0 years", "5.1-10.0 years", "10.1-15.0 years", "15.1+ years"], zero_division= 0)
print(class_report_train)

comb_line_test_Y_clf = comb_line_test_Y.apply(classify)
comb_line_pred_test_Y = [classify(y) for y in comb_line_pred_test_Y]
class_report_test = classification_report(comb_line_test_Y_clf, comb_line_pred_test_Y, target_names=["0.0-5.0 years", "5.1-10.0 years", "10.1-15.0 years", "15.1+ years"], zero_division= 0)
print(class_report_test)
```

```python
variable_names = comb_line_X.columns
weights_linear = comb_line_clf.coef_[0]
for varr, weighttt in zip(variable_names, weights_linear):
    print("Variable: ", varr, ", Weight: ", weighttt)
```

```python
def display_weights(model, variables):
    weights = model.coef_
    weights_df = pd.DataFrame({'Variables': variables, 'Weight': weights})
    weights_df['Weight Absolute Value'] = weights_df['Weight'].abs()
    weights_df = weights_df.sort_values(by = 'Weight Absolute Value', ascending = False)
    weights_df = weights_df.drop(columns = ['Weight Absolute Value'])
    print("\nVariables with the most influence:\n")
    for index, row in weights_df.iterrows():
        print("Variable: ", row['Variables'], ", Weight: ", row['Weight'])


comb_X = df_cleaned[['WS', 'TS%', 'PER', 'BPM', 'DWS', 'VORP', 'MP', 'TOV%']]
comb_Y = df_cleaned['career_length']

comb_train_X, comb_test_X, comb_train_Y, comb_test_Y = train_test_split(
    comb_X,
    comb_Y,
    test_size = 0.2,
    random_state = 42
)

scaler = StandardScaler()
comb_train_X = scaler.fit_transform(comb_train_X)
comb_test_X = scaler.transform(comb_test_X)

comb_ridge = Ridge(alpha = 0.1)
comb_ridge.fit(comb_train_X, comb_train_Y)

comb_train_pred = comb_ridge.predict(comb_train_X)
comb_test_pred = comb_ridge.predict(comb_test_X)

mse_train_comb = mean_squared_error(comb_train_Y, comb_train_pred)
mse_test_comb = mean_squared_error(comb_test_Y, comb_test_pred)
r2 = r2_score(comb_test_Y, comb_test_pred)
print("Training mean squared error: ", mse_train_comb, "Testing mean squared error: ", mse_test_comb, "R squared: ", r2)

comb_train_Y_clf = comb_train_Y.apply(classify)
comb_train_pred_clf = [classify(y) for y in comb_train_pred]
class_report_train = classification_report(comb_train_Y_clf, comb_train_pred_clf, target_names=["0.0-5.0 years", "5.1-10.0 years", "10.1-15.0 years", "15.1+ years"], zero_division= 0)
print(class_report_train)

comb_test_Y_clf = comb_test_Y.apply(classify)
comb_test_pred_clf = [classify(y) for y in comb_test_pred]
class_report_test = classification_report(comb_test_Y_clf, comb_test_pred_clf, target_names=["0.0-5.0 years", "5.1-10.0 years", "10.1-15.0 years", "15.1+ years"], zero_division= 0)
print(class_report_test)
```

# Ethics & Privacy

### Privacy of player data and injury information 
Although our study relies on publicly available NBA player statistics based on the last 20 seasons, privacy concerns might arise on the topic of dealing with specific player information, such as injury history and external factors that might influence career longevity. Michael Beasley is a prominent example of this. Although he showed flashes of brilliance early on in his career, his lack of professionalism and off-court issues led to a shorter than expected NBA career. While our study is mainly focusing on data relating to individual player performance, it is important to note that career longevity is affected by many variables off the court, such as a player’s personal choices as well as outside factors that are out of their control. These external factors, such as lifestyle and personal history, cannot be directly incorporated into the scope of our analysis. With player injury information in particular, it can be difficult to uncover the specific details of the injury histories released by teams because they are relatively vague. It is difficult to discern the impact of these injuries on specific players by accessing reports online alone, so this brings up the question of if incorporating injury data is a reasonable choice or not. Overall, it is important to balance the need for comprehensive data analysis with respecting the individual privacy of each player, especially when dealing with the long-term impacts of how injuries can affect a player’s career.

### Bias in statistical metrics favoring offensive performance 
Many statistical metrics used in the NBA like player efficiency rating, box-plus-minus, and fantasy score, are inherently biased towards certain archetypes or playstyles. In particular, offensive contributions are far more quantifiable than defensive contributions. While a great offensive player’s impact is clearly shown in their statline, a strong defensive player’s influence on the game is not as evident from metrics alone. This inherent skew in importance is a direct result of how these statistics are recorded. From a defensive perspective, the number of blocks a defender gets per game is not directly correlated with their overall defensive performance. That is to say, a player can have a low number of blocks recorded but might be a great defender overall. This disparity between offensive and defensive contributions needs to be carefully handled, and required a delicate balance of which metrics to include in our analysis.

### Basic mitigation strategies for these issues 
To address privacy issues relating to player data, we made sure to strictly limit our analysis to publicly available statistics, and also acknowledged the limitations of our study by noting that statistical metrics alone cannot fully explain career longevity. By excluding external factors that might otherwise be difficult to quantify such as ones that relate with a player’s personal life off the court, our analysis becomes more generalizable while simultaneously evading privacy issues. In the same vein, it was in our best interest to exclude injury histories as a variable in our analysis because public injury reports might not cover the full context around a player’s injury. Additionally, it is known that teams occasionally list players on the injury report to let certain players get rest during games to prevent injuries, which is colloquially known as load management. This further supports not directly using injury histories as a metric in our data. 

During analysis, we effectively mitigated bias by incorporating a wide variety of metrics to deter the influence of spurious correlations and confounding variables. Since offensive contributions are often favored in the context of NBA statistics, we attempted to balance our metrics by including a meaningful combination of offensive, defensive, and more advanced metrics. Some examples of defensive metrics we tried to utilize included box plus minus and defensive win shares. Creating a well-defined balance of metrics negates the presence of bias overall as it allows for a nuanced and standardized analysis of players of vastly different playstyles and archetypes. We implemented separate analyses for players of different positions as well, such as placing guards, forwards, and centers in different groups to analyze. 

Furthermore, we should emphasize that our findings are merely for descriptive and exploratory purposes and should not be used as a strict set of guidelines to adhere to, and that further research should be encouraged to look at a multitude of views. Teams should consider a wide range of factors when deciding on which players to recruit, including many factors that cannot be quantified via our model. It is important to note that our model’s intent is not to be used to predict a player’s future, but rather, to examine preexisting trends based on prior data. Ultimately, this project is one that seeks to understand and uncover past trends, rather than serving as a definitive predictor of player longevity.

## Discussion

Our analysis aimed to determine which advanced statistical metrics best predict the longevity of an NBA player’s career across different positions. Through data analysis and model testing, we analyzed various metrics, including Box Plus-Minus (BPM), Win Shares (WS), Defensive Win Shares (DWS), and True Shooting Percentage (TS%), to determine their predictive power.

When analyzing positional differences, the results showed that BPM had the highest predictive weight and a positive correlation with career length, indicating that well-rounded players who contribute across multiple aspects of the game tend to have longer careers. Win Shares (WS) was the second most important factor, reinforcing the idea that long-term contributions to team success impact longevity. Additionally, Minutes Played and Defensive Win Shares (DWS) exhibited strong correlations with career length, suggesting that durability and defensive value contribute to extended careers.

Interestingly, for guards, True Shooting Percentage (TS%) ranked as the fifth most important variable but had a negative correlation with career length. This suggests that while scoring efficiency is valuable, it does not necessarily lead to extended careers, potentially due to factors such as defensive ability, adaptability to different roles, and the physical demands placed on guards. Meanwhile, forwards and centers showed a stronger correlation between Win Shares and longevity, indicating that their contributions in rebounding, defense, and overall team success are key to sustaining long careers.

Despite these findings, our analysis is limited by the absence of injury data, which is a crucial factor in career sustainability. Future research that includes injury history, player workload, and team performance metrics could provide a more comprehensive understanding of career longevity predictors.

Our study illustrates the importance of Box Plus-Minus (BPM) as the strongest predictor of NBA career longevity, highlighting the value of well-rounded contributions beyond scoring. Win Shares (WS) remained a key factor, though its significance varied across positions, particularly for centers where it was far less predictive. True Shooting Percentage (TS%) showed a surprising negative correlation with career length, suggesting that efficiency alone does not guarantee longevity, particularly for guards.

These results challenge the conventional view that scoring efficiency is the primary indicator of sustained success. Instead, a player’s ability to impact multiple facets of the game, especially on defense and within the team’s overall performance, appears to be a more accurate reflection of long-term career viability.

Future research should incorporate more sophisticated models, including machine learning techniques, expanded datasets, and an examination of external factors such as injuries and team dynamics.
